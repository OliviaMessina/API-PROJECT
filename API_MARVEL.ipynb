### TABLE URL

```python
import pyodbc 
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=DESKTOP-QVPS0TH;'
                      'Database=MARVEL;'
                      'Trusted_Connection=yes;')
cursor = conn.cursor()
sql = """CREATE TABLE URL(
        Id_URL int PRIMARY KEY,
        Id_Character int,
        Url varchar(8000),
        CONSTRAINT FK_id_character_url1 FOREIGN KEY (Id_Character)
        REFERENCES Characters(Id_Character));"""

cursor.execute(sql)

print("reussi")
cursor.commit()
```
